## Is climate change eating away at Tromsø's skiing season?
This notebook is  for retrieving and analysing data from the [Frost API](https://frost.met.no/index.html) made available by *Meteorologisk institutt*, the Norwegian weather organisation. It is also for learning along the way.

I want to find out if there is less snow in Tromsø, a town of 80,000 people in the Arctic part of Norway - and my home - than there used to be.

##### Some notes about the snow depth data

The [documentation](https://frost.met.no/concepts2.html#relationshipreftime) says this:
"TimeResolution is the period between each data value, i.e. data output frequency. 'timeResolution': 'P1D' means one data value once a day." 
*and* 
"TimeOffset is observation time relative to midnight. 'timeOffset': 'PT6H' means that the daily data value has observation time 06:00 UTC"

This means that my snow depth data has one data value per day (because `'timeResolution' == 'P1D'`, and that the snow depth is observed at 6 am (because `'timeOffset' == 'PT6H'`.

The observed snow depth data is therefore neither a mean or a max like I thought earlier, but an observation made at 6 am in the morning (which is also what I've heard as a journalist when talking to the people working at this weather station).
